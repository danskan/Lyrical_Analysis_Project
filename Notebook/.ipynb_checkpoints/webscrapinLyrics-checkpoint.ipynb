```python
song_artist = df[['song', 'artist']]
song_artist.head()
```

```python
# create a list of urls for artists

# from azapi import AZlyrics

artist_list = song_artist["artist"].to_list()
# api = AZlyrics()
artist_url_list = []


for i in range(len(artist_list)):
    artist_lower_case = artist_list[i].replace(' ','').lower()
    first_letter = artist_list[i][0].lower()
    url = f"https://www.azlyrics.com/{first_letter}/{artist_lower_case}.html"
    artist_url_list.append(url)

print(f"Pre dedup: {len(artist_url_list)}")
artist_url_list = set(artist_url_list)
list(artist_url_list)
print(f"Post dedup: {len(artist_url_list)}")
# print(f"Pre dedup: {artist_url_list}")
# artist_url_list = list(set(artist_url_list))
# print(artist_url_list)
# print(len(f"Length of deduped file: {artist_url_list}"))
```

```python
# request for the html response using the url for wikipedia
# to scrape the page to see if we get value 200 in return, confirming acceptable to utilize page info

lyric_url_list = []
counterK = 0


for i in range(len(artist_url_list)):
    url = artist_url_list[i]
    
    # url=artist_url_list[i]
#     response = requests.get(url)
    
    # table_class="wikitable sortable plainrowheaders jquery-tablesorter"
    response=requests.get(url)
    
    # print(response.status_code)
    soup = BeautifulSoup(response.text, 'html.parser')
    songList=soup.find_all('div', {'class':"listalbum-item"})
    
    print(f"{artist_url_list[i]}: {i} of {len(artist_url_list)}")
    
    counterK += 1
        
    if counterK % 75 == 0:
        print("going to nap for a bit")
        snooze = random.randint(10,30)
        time.sleep(snooze)
        counterK = 0
    
    counterJ = 0
    
    for j in range(len(songList)):
        
        try:

            data = songList[j].find('a')['href']
            lyric_url = f"https://www.azlyrics.com{data}"
            lyric_url_list.append(lyric_url)

            print(f"{data}: {j} of {len(songList)}")

            counterJ += 1
        
        except TypeError:
            print("TypeError: 'NoneType' object is not subscriptable")
        
        if counterJ % 75 == 0:
            print("going to nap for a bit")
            snooze = random.randint(10,30)
            time.sleep(snooze)
            counterJ = 0


```
