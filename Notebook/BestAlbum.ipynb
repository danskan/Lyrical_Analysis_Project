```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# request for the html response using the url for wikipedia
# to scrape the page to see if we get value 200 in return, confirming acceptable to utilize page info

url="https://en.wikipedia.org/wiki/List_of_best-selling_albums"
response = requests.get(url)
# table_class="wikitable sortable jquery-tablesorter"
response=requests.get(url)
print(response.status_code)
```

```python
# parse data from the html into a beautifulsoup object
soup = BeautifulSoup(response.text, 'html.parser')
albumTable=soup.find_all('table',{'class':"wikitable sortable"})
albumTable
```

```python
# grab the tags with table titles using beautiful soup
title_list = soup.find_all('span',{'class':"mw-headline"})
title_list
```

```python
# create clean list of table titles
# *may not be leveraged
for title in title_list:
  title_clean = title.text
  print(title_clean)
```

```python
# create a df from the album table
# loop through, using range to limit observations to desired portions of url
# use pd.concat to append df row to df
df=pd.read_html(str(albumTable))

# create empty df to allow population in loop
best_selling_df = pd.DataFrame()

for i in range(len(albumTable)-2):
  if 0:
    best_selling_df = df[i]
  else:
    best_selling_df = pd.concat([best_selling_df, df[i]])

print(best_selling_df.head())
print(best_selling_df.tail())
print(best_selling_df.columns)
```

```python
# iterate through columns using map function to change values to lower case
best_selling_df.columns = map(str.lower, best_selling_df.columns)
best_selling_df.columns
```

```python
# push to a csv
best_selling_df.to_csv('bestsellingalbums.csv', encoding='utf-8', index=False)
```
